```python
import matplotlib.pyplot as plt


#function to multiply the matrix and the vector
def mul_fun(x,y):
  if len(x[0])!=len(y):
    raise ValueError("the no of column in matrix and vector lenght should be equal")
  result= []
  for i in range(len(x)):
        sum = 0
        for k in range(len(y)):
            sum += x[i][k] * y[k] #multiplay the elements of the matrix and the vector
        result.append(sum)
  return result


#function to plot the line graph
def plot_graph(A, u, v):
    #defining start and end points
    start1 = (0, 0)
    end1 = mul_fun(A, u)

    start2 = (0, 0)
    end2 = mul_fun(A, v)

    start3 = (0, 0)
    end3 = u

    start4 = (0, 0)
    end4 = v

    # Create a figure and axis
    fig, ax = plt.subplots()

    # Plot lines
    #give strat,end,label and color for the vector A*u
    ax.plot([start1[0], end1[0]], [start1[1], end1[1]], color='blue', label='A*u')

    #give strat,end,label and color for the vector A*v
    ax.plot([start2[0], end2[0]], [start2[1], end2[1]], color='red', label='A*v')

    #give strat,end,label and color for the vector u and linestyle to diffenciate
    ax.plot([start3[0], end3[0]], [start3[1], end3[1]], color='yellow', linestyle='--', label='u')

    #give strat,end,label and color for the vector v and linestyle to diffenciate
    ax.plot([start4[0], end4[0]], [start4[1], end4[1]], color='green', linestyle='--', label='v')

    # Graph labeling of x and y axis
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Visualization of Eigenvector')
    plt.axis('equal')
    plt.legend()
    plt.grid(True)
    plt.show()



#define the matrix and vector
A =[[1,6],[5,2]]
u=[6,-5]
v=[3,-2]
plot_graph(A,u,v)
```
